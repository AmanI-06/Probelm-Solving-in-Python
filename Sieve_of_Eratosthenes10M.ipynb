```
def SieveOfEratosthenes10M(num): 
      
    
    PrimeArr = [True for i in range(num+1)] 
    n= 2
    while (n * n <= num): 
          
        
        if (PrimeArr[n] == True): 
              
            
            for i in range(n * n, num+1, n): 
                PrimeArr[i] = False
        n += 1
      
  
    for n in range(2, num+1): 
        if PrimeArr[n]: 
            print (n)

if __name__=='__main__': 
    num = 10000000
    print ("The prime numbers upto",num )
    SieveOfEratosthenes10M(num) 
```
